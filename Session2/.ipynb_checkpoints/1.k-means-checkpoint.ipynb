### 1.라이브러리 로딩

```python
# pandas 라이브러리
import pandas as pd
# 머신러닝 라이브러리
from sklearn.cluster import KMeans
# 시각화 라이브러리
import matplotlib.pyplot as plt
```

```python
# 시각화 inline으로 출력
%matplotlib inline
```

### 2. 데이터셋 불러오기

```python
irisData = pd.read_csv("../dataset/IRIS.csv")
irisData.head(5)
```

```python
species = irisData[['SPECIES']]
species.head(5)
```

```python
features = irisData[[ 'PETALLENGTH', 'PETALWIDTH']]
features.head(5)
```

### 3. 모델구동(K-means)

```python
model= KMeans(n_clusters=3,algorithm='auto') #, algorithm="auto"
model.fit(features)
predict = pd.DataFrame(model.predict(features))
predict.columns=['PREDICT_K']
predict.head(5)
```

### 4.결과 데이터 출력( csv파일로 저장)

```python
r=pd.concat([species,features,predict],axis=1)
r
```

```python
r.to_csv("d:/iris_result1.csv")
```

### 5. 결과데이터 시각화

```python
plt.scatter(r['PETALLENGTH'],r['PETALWIDTH'],c=r['PREDICT_K'],alpha=0.7)
plt.show()
```
